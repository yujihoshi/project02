Outliers are an interesting problem in statistics, in that there is not an agreed upon best way to define them. Subjectivity in selecting and analyzing data is a problem that will recur throughout the course.

1. Pull out the rate variable from the sat dataset.
2. Are there outliers in the dataset? Define, in words, how you _numerically define outliers._

````python
sat_df['Rate']
````

Let us Define Outliers as 1.5x standard deviations away from the mean

````python
sat_df['Rate'].mean()
````

````python
sat_df['Rate'].std()
````

````python
upper_limit = sat_df['Rate'].mean()+1.5*sat_df['Rate'].std()
````

````python
lower_limit = sat_df['Rate'].mean()-1.5*sat_df['Rate'].std()
````

By our definition, anything above 78.1 is an outlier.

````python
outliers = sat_df[sat_df['Rate']> upper_limit]['Rate']
outliers
````

````python
new_rate = sat_df[sat_df['Rate']< upper_limit]['Rate']
````

````python
print(new_rate.mean(), new_rate.std(), new_rate.median())
````

````python
print(sat_df['Rate'].mean(), sat_df['Rate'].std(), sat_df['Rate'].median())
````

### 9.1 Calculate the spearman correlation of sat `Verbal` and `Math`

1. How does the spearman correlation compare to the pearson correlation? 
2. Describe clearly in words the process of calculating the spearman rank correlation.
  - Hint: the word "rank" is in the name of the process for a reason!


````python
stats.spearmanr(sat_df['Math'],sat_df['Verbal'])
````

````python
sat_df.corr(method = 'spearman')
````

````python
stats.pearsonr(sat_df['Math'],sat_df['Verbal'])
````

Spearman Rank is ranked 1 when similar, and -1 when dissimilar.
Rather than looking for linear relationships, it looks for monotonic relationships.
ie (1) as the value of one variable increases, so does the value of the other variable; or (2) as the value of one variable increases, the other variable value decreases.
